Lambda Functions

```python
def raiseToThePowerOfTwo(x):
    return x ** 2
```

```python
raiseToThePowerOfTwo(3)
```

```python
raiseToThePowerOfTwoLambda = lambda x: x ** 2
```

```python
raiseToThePowerOfTwoLambda(5)
```

```python
(lambda x: x / 2) (11)
```

```python
sum_xy = lambda x, y: x + y
```

```python
sum_xy(4, 8)
```

```python
sum_XY = lambda x, y: x + y(x)
```

```python
sum_XY(3, lambda x: x * 5)
```
